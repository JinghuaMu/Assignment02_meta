```r
# Convert the equation to be solved into the form f(V,ci) = 0
func1 <- function(V,VM,F,K,Ci){
   exp((V-VM)/(F*K)) + V/(F*Ci) - 1
}

# Since our transformed function f increases with V and ci, it can be solved iteratively
#define a function to solve the root of monotonically increasing function
iter_func <- function(fun,ini,step,...){
   y = fun(ini,...)
   current = ini
   while (y < 0){
      current = current + step
      y = fun(current,...)
   }
   current - step/2
}


# the initial parameters and bounds of ci
Ci_range = seq(0.001,3,0.01)
VM_c = 1
F_c = 1
K_c = 0.1

# giving ci to solve specific V
V_range = c()
V_range[1] = iter_func(func1,0,0.001,VM_c,F_c,K_c,Ci_range[1])
j = 2
for (i in Ci_range[-1]) {
V_range[j] = iter_func(func1,V_range[j-1],0.0001,VM_c,F_c,K_c,i)
j = j + 1
}

# Estimated value of V in giving ci
V_e1 = VM_c*Ci_range/(Ci_range + K_c)
V_e2 = F_c*Ci_range

#Red line : Estimated value of V at Vmax > V (flow-limited)
#Blue line: Estimated value of V at Vmax >> V (homogeneous)
#Vertical line: The transition ci* between the homogeneous and flow-limited regimes
#bold line: Estimated value using iterative root finding
plot(x = Ci_range, y = V_range, type = 'l', lwd = 3, 
    ,main = "V in equilibrium change with Ci ", xlab = "initial concentration" ,ylab = "Total rate of removal in equilibrium")
lines(x = Ci_range, y = V_range
lines(x = Ci_range, y = V_e1,col = "red",lty=2, lwd = 2)
lines(x = Ci_range, y = V_e2, col = "blue",lty = 2, lwd = 2)
abline(v = VM_c/F_c - K_c, col = "purple")
text(2.2, 0.1, expression(homogeneous: hat(V) == (V[MAX]*c[i])/(c[i]+K)), col = "blue")
text(2, 0.2, expression(flow-limited: hat(V) == F/c[i]), col = "red")
text(VM_c/F_c - K_c, 0.1, expression("transition c"[i]), col = "purple")
```
